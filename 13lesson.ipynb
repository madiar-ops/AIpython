```python
import torch
import torch.nn as nn
from transformers import pipeline

#часть 1

print(f"PyTorch version: {torch.__version__}")

gpu_available = torch.cuda.is_available()
print(f"GPU available: {gpu_available}")
if gpu_available:
    print(f"GPU name: {torch.cuda.get_device_name(0)}")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Используемое устройство: {device}")

tensor = torch.rand(3, 4, dtype=torch.float32)
print(tensor)

print(f"Shape: {tensor.shape}")
print(f"Dtype: {tensor.dtype}")
print(f"Mean:  {tensor.mean().item():.4f}")
print(f"Sum:   {tensor.sum().item():.4f}")

#часть 2

model = nn.Sequential(
    nn.Linear(4, 8),
    nn.ReLU(),
    nn.Linear(8, 3)
)
print(model)

X = torch.tensor([
    [5.1, 3.5, 1.4, 0.2],
    [6.2, 3.4, 5.4, 2.3],
    [5.9, 3.0, 4.2, 1.5]
], dtype=torch.float32)
print(X)

output = model(X)
print("Output:")
print(output)
print("Output shape:", output.shape)

predictions = torch.argmax(output, dim=1)
print("Predictions:", predictions)


#часть 3

classifier = pipeline("sentiment-analysis")

reviews = [
    "This product completely exceeded my expectations, I love it!",
    "Worst purchase I have ever made, total waste of money.",
    "The quality is decent, but the price is a bit too high.",
    "Absolutely fantastic service, the staff were incredibly friendly.",
    "I am very disappointed, the item broke after just two days.",
    "Great value for money, I would definitely buy this again.",
    "The delivery was late and the packaging was damaged.",
    "An amazing experience overall, highly recommend it to everyone.",
    "It's okay, nothing special but it does the job.",
    "Terrible customer support, nobody answered my questions."
]

for review in reviews:
    result = classifier(review)[0]
    print(f"Text:  {review}")
    print(f"Label: {result['label']}")
    print(f"Score: {result['score']:.4f}")
    print("-" * 60)

#часть 4

def analyze_review(text):
    """
    Анализирует тональность одного отзыва с помощью classifier.

    Параметры:
        text (str): текст отзыва

    Возвращает:
        dict: {"text": str, "label": str, "score": float}
    """
    result = classifier(text)[0]
    return {
        "text": text,
        "label": result["label"],
        "score": float(result["score"])
    }

new_reviews = [
    "The new update made the app so much faster and easier to use.",
    "I regret buying this, it stopped working after one week.",
    "Pretty average experience, nothing to complain about.",
    "Best customer service I have ever encountered, truly outstanding.",
    "The food was cold and the order was completely wrong."
]

for text in new_reviews:
    print(analyze_review(text))
```
