```python
import pandas as pd

# ---------- Задача 25 ----------
data = {
    "Имя": ["Али", "Маша", "Даня", "Сара", "Илья", "Оля", "Тимур", "Лена"],
    "Возраст": [18, 19, 18, 20, 21, 19, 22, 20],
    "Оценка": [85, 90, 78, 92, 88, 76, 95, 89],
    "Город": ["Алматы", "Астана", "Алматы", "Шымкент", "Астана", "Алматы", "Шымкент", "Астана"]
}

df = pd.DataFrame(data)

print("Исходная таблица:")
print(df)

# ---------- Задача 26----------

print("\nПервые 5 строк:")
print(df.head())

print("\nНазвания столбцов:")
print(df.columns)

print("\nТолько имя и оценка:")
print(df[["Имя", "Оценка"]])

print("\nПервая строка:")
print(df.iloc[0])

print("\nСтуденты с оценкой > 85:")
print(df[df["Оценка"] > 85])

print("\nСортировка по оценке:")
print(df.sort_values(by="Оценка", ascending=False))

df["Статус"] = df["Оценка"].apply(lambda x: "Сдал" if x >= 80 else "Не сдал")
print("\nС добавленным столбцом:")
print(df)

print("\nСредний балл:")
print(df["Оценка"].mean())

print("\nСредний балл по городам:")
print(df.groupby("Город")["Оценка"].mean())

# ---------- Задача 27 ----------
print("\nВыводы:")
print("1. Средний балл студентов находится на хорошем уровне.")
print("2. Самые высокие оценки у студентов из Шымкента.")
print("3. Большинство студентов успешно сдали.")
print("4. Есть студенты с более низкими результатами, которым нужно подтянуть знания.")
```
