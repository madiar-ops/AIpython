```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


ads = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
sales = np.array([100, 130, 160, 190, 220])

model1 = LinearRegression()
model1.fit(ads, sales)

sleep_hours = np.array([4, 5, 6, 7, 8]).reshape(-1, 1)
productivity = np.array([45, 52, 61, 72, 82])

model2 = LinearRegression()
model2.fit(sleep_hours, productivity)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(ads, sales, color='blue', label='Данные')
plt.plot(ads, model1.predict(ads), color='red', label='Регрессия')
plt.title('Реклама vs Продажи')
plt.xlabel('Реклама')
plt.ylabel('Продажи')
plt.legend()

#  2
plt.subplot(1, 2, 2)
plt.scatter(sleep_hours, productivity, color='green', label='Данные')
plt.plot(sleep_hours, model2.predict(sleep_hours), color='orange', label='Регрессия')
plt.title('Сон vs Продуктивность')
plt.xlabel('Часы сна')
plt.ylabel('Продуктивность')
plt.legend()

plt.tight_layout()
plt.show()

print(f"Практика 1: Прогноз продаж = {model1.predict([[60]])[0]:.2f} при 60 ед. рекламы")
print(f"Практика 2: Прогноз продуктивности = {model2.predict([[9]])[0]:.2f} при 9 часах сна")
```
